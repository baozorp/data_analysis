```python
import random
import shutil
import torch
import yaml
import os

from pathlib import Path

from ultralytics import YOLO

from sklearn.model_selection import train_test_split
```

```python
def train_yolo_model(main_experiment: str, name: str, seed: int):
    model = YOLO("yolov9e-seg.pt")

    ABS_PATH = os.getcwd()
    results = model.train(
        data = os.path.join('/kaggle/input/itas-2024', 'data', main_experiment, name, 'data.yaml'),
        project = 'results',
        name = name,
        epochs = 25,
        patience = 0,
        batch = 5,
        imgsz = 640,
        seed=seed
    )
```

```python
for i in range(5):
    train_yolo_model(main_experiment='train_set', name=f'1_train', seed=(i + 1) * 1)
```

```python
def least_confidence(top_percentage=100, confidence_threshold=0.1):
    model = YOLO(Path('/kaggle/input/itas-2024/data/best.pt'))

    paths_to_images = []    # Ğ°Ğ±ÑĞ¾Ğ»ÑÑ‚Ğ½Ñ‹Ğµ Ğ¿ÑƒÑ‚Ğ¸ Ğ´Ğ¾ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ğ¹
    for dirname, _, filenames in os.walk('/kaggle/input/itas-2024/data/dataset/images'):
        for filename in filenames:
            if filename.endswith('.jpg'):
                paths_to_images.append(Path(dirname) / Path(filename))

    selected_samples = []

    for i, path_to_image in enumerate(paths_to_images):
        print(f'{i}/{len(paths_to_images)} ->> {path_to_image}')
        results = model.predict(path_to_image, conf=confidence_threshold)
        result = results[0]

        selected_samples.append({'image_path': result.path, 'min_confidence': min(result.boxes.conf), 'result': result})
    
    selected_samples.sort(key=lambda x: x['min_confidence'])
    selected_samples = selected_samples[:int(len(paths_to_images) * (top_percentage / 100))]
    
    return selected_samples
```

margin_sampling

```python
def margin_sampling(top_percentage=100, confidence_threshold=0.01):
    model = YOLO(Path('/kaggle/input/itas-2024/data/best.pt'))

    paths_to_images = []    # Ğ°Ğ±ÑĞ¾Ğ»ÑÑ‚Ğ½Ñ‹Ğµ Ğ¿ÑƒÑ‚Ğ¸ Ğ´Ğ¾ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ğ¹
    for dirname, _, filenames in os.walk('/kaggle/input/itas-2024/data/dataset/images'):
        for filename in filenames:
            if filename.endswith('.jpg'):
                paths_to_images.append(Path(dirname) / Path(filename))

    selected_samples = []

    for i, path_to_image in enumerate(paths_to_images):
        results = model.predict(path_to_image, conf=confidence_threshold)
        result = results[0]

        classes = result.boxes.cls
        
        if (4. in classes and 5. in classes):
            print(f'{i}/{len(paths_to_images)} True ->> {path_to_image}')
            
            results_4_class = result[[i for i, cls in enumerate(classes) if cls in [4.]]]
            results_5_class = result[[i for i, cls in enumerate(classes) if cls in [5.]]]

            min_conf_4_class = min(list(results_4_class.boxes.conf))
            max_conf_4_class = max(list(results_4_class.boxes.conf))

            min_conf_5_class = min(list(results_5_class.boxes.conf))
            max_conf_5_class = max(list(results_5_class.boxes.conf))

            if abs(max_conf_5_class - min_conf_4_class) > abs(max_conf_4_class - min_conf_5_class):
                selected_samples.append(
                    {'image_path': result.path, 'max_margin': abs(max_conf_5_class - min_conf_4_class)})
            else:
                selected_samples.append(
                    {'image_path': result.path, 'max_margin': abs(max_conf_4_class - min_conf_5_class)})
        else:
            print(f'{i}/{len(paths_to_images)} False ->> {path_to_image}')
    selected_samples.sort(key=lambda x: x['max_margin'])
    selected_samples = selected_samples[:int(len(paths_to_images) * (top_percentage / 100))]
    
    return selected_samples
```

```python
samples = margin_sampling()
```
